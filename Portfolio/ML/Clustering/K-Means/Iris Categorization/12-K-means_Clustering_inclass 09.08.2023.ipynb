## Hopkins test

Hopkins testi, veri kmesinin dzgn (uniform) bir dalma sahip olup olmadn tahmin etmek iin kullanlan bir istatistiksel testtir. Clustering Random data ister.

Hopkins skoru 0'a yaklayorsa, bu, veri kmesinin dzgn bir dalma sahip **olmad** anlamna gelir. Bu durumda, veriyi snflandrmak iin kmeleme (clustering) yntemleri kullanmak yararl olabilir. (clustering e uygunh demek). 

Ancak Hopkins skoru belirli bir deerin (rnein 0.3) zerindeyse, bu, veri kmesinin dzgn bir dalma sahip olduu anlamna gelir. Bu tr bir durumda, kmeleme yntemlerini kullanmak pek anlaml olmayabilir. Literatrde 0.5 diyenlerde var hatta bak gibi 0.55 uygun deil dememek lazm. Best Practice.

zetle, Hopkins testi, bir veri kmesinin dzgn bir dalm gsterip gstermediini ve bu balamda kmelemenin uygun olup olmadn deerlendirmenizi salar.

```python
## HOPKINS TEST N MANUEL KOD

from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd
def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    # Get the distance to their neirest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    # Get the distance to their neirest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]


# normalde bu fonksiyonumuz pyclustertend'n hopkins ktphanesinde var. Ancak oluabilecek hatalara kar 
# dorudan fonksiyonun kendisini kullanacaz. Bundan dolay pyclustertend' install etmemize gerek yoktur! 
                                                #!pip install pyclustertend --user
                                                # from pyclustertend import hopkins
```

# Choosing The Optimal Number of Clusters

## Elbow metod

**Inertia (Kmeler i Toplam Kareler):** Inertia, her veri noktasnn kme merkezine olan uzaklnn karesinin toplamdr. Dk bir inertia deeri, veri noktalarnn kendi kme merkezlerine daha yakn olduunu gsterir, bu da genellikle iyi bir kmeleme olduunu dndrebilir. Ancak, inertia deeri, kmelerin says arttka genellikle azalr. Bu nedenle, farkl kme saylar iin inertia deerlerini karlatrarak en uygun kme saysn semek zordur.

```python
ssd = [] ##inertia

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 
plt.show()
```

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff() # sadece DataFrame serieste alr. index srasna gre satrlarn farkn alr. Bu skorlara oransal olarak baklmal. Aksi halde yanlrz
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

## Silhouette analysis

Silhouette analizi, kmeleme performansn deerlendirmek iin kullanlan bir metriktir. Kme ii benzerlii ve kmeler aras benzerlii karlatrarak, kmelerin ne kadar iyi ayrldn gsterir. Silhouette deeri, her veri noktas iin hesaplanr ve bu deerlerin ortalamas alnarak genel bir kmeleme kalitesi lm elde edilir.

Silhouette deeri -1 ile 1 arasnda deiir:

1: i veri noktas kendi kmesinde iyi yerletirilmi ve dier kmelerden uzakta demektir.

0: i veri noktas iki kme arasnda bir snra yakn demektir.

1: i veri noktas yanl bir kmeye atanm demektir.

bu deerlerin "eik" deerleri kesin bir ekilde belirlenmemitir. Ancak, pratikte, ortalama silhouette deerinin yksek olmas (genellikle 0.5'ten byk) kmelenmenin iyi olduunu gsterirken, dk deerler (genellikle 0.25'ten dk) kmelenmenin iyi olmadn veya daha iyi yaplabileceini gsterebilir.

Silhouette, veri kmeleri iindeki tutarlln yorumlanmas ve dorulanmas yntemini ifade eder. Teknik, her bir nesnenin ne kadar iyi snflandrldnn ksa ve z bir grafik temsilini salar. Siluet deeri, bir nesnenin dier kmelere kyasla kendi kmesine ne kadar benzediinin bir lsdr.

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, K_means_model.labels_)
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

Siluet Skoru: Grselde her bir veri noktas iin bir siluet deeri (katsay) vardr. Bu deerler -1 ile 1 arasnda deiir. 1'e yakn deerler iyi bir kmelemeyi, 0'a yakn deerler bir kmse snrna yakn olan veri noktalarn ve -1'e yakn deerler ise yanl bir kmseye atanm veri noktalarn gsterir.

Kmelerin Boyutu: Grselde her kme iin bir siluet grafii vardr. Bu grafikler, kmelerin ne kadar homojen olduunu ve kmeler arasndaki ayrmn ne kadar belirgin olduunu gsterir.

Krmz Noktal izgi: Bu izgi, tm veri noktalar iin ortalama siluet skorunu gsterir. Bu izginin zerindeki kmeler genellikle iyi bir performans gsterirken, bu izginin altndaki kmeler daha dk bir performans gsterir.

Kme Genilikleri: Her bir kme iin siluet genilii, o kmeye ait veri noktalarnn siluet katsaylarn gsterir. Eer tm kmeler benzer geniliklere sahipse, bu, kmelerin benzer younlukta ve boyutta olduunu gsterir. Farkl geniliklerde kmeler, kmelerin boyut ve younluklarnn farkl olduunu gsterir.

SilhouetteVisualizer' yorumlarken dikkat edilmesi gerekenler:

- Geni siluetler (1'e yakn deerler) iyi bir kmeleme performansn gsterir.

- Dar veya negatif deerlere sahip siluetler, bu veri noktalarnn muhtemelen yanl kmeye atandn gsterir.

- Eer birok veri noktas ortalama siluet skorunun (krmz noktal izgi) altndaysa, bu kmse saysnn gzden geirilmesi gerektiini gsterebilir.

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=3, random_state=42)
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```

## Adjust Rand Score

Rand ndeksi, tm rnek iftlerini gz nnde bulundurarak ve tahmin edilen ve gerek kmelemelerde ayn veya farkl kmelere atanan iftleri sayarak iki kmeleme arasnda bir benzerlik ls hesaplar. Cezanladrma gibi bir hesab vardr.

Rand indeksi, 0 ile 1 arasnda bir deere sahiptir; 0, iki veri kmelemesinin herhangi bir nokta ifti zerinde uyumadn ve 1, veri kmelemelerinin tamamen ayn olduunu belirtir.
